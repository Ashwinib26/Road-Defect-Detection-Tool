### Computer Vision Concept - Morphological Operations  (Dilation, Erosion, Opening, Closing and Gradient Display of an image)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_img = "your_image/path"
image = cv2.imread(original_img, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image)
```

```python
_ , binary_img = cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)               #binary thresholding (inverse here)
```

```python
kernel = np.ones((5,5),np.uint8)  #kernel is basically a 'filter' that movess across the img to get the desired changes
```

```python
# dilation , erosion 'opening' and 'closing' is used in object detection and all
```

```python
dilated_img = cv2.dilate(binary_img, kernel, iterations=1)
eroded_img = cv2.erode(binary_img, kernel, iterations=1)
opened_img = cv2.morphologyEx(binary_img, cv2.MORPH_OPEN, kernel)
closed_img = cv2.morphologyEx(binary_img, cv2.MORPH_CLOSE, kernel)
gradient_img = cv2.morphologyEx(binary_img, cv2.MORPH_GRADIENT, kernel)
```

```python
images = [image, dilated_img, eroded_img, opened_img, closed_img, gradient_img]
titles = ["Original", "Dilated", "Eroded", "Opened", "Closed", "Gradient"]
```

```python
plt.figure(figsize=(10, 8))
for i, (img, title) in enumerate(zip(images, titles)):
    plt.subplot(2, 3, i + 1)
    plt.imshow(img, cmap='gray')  # Ensure grayscale display
    plt.axis('off')
    plt.title(title)

plt.tight_layout()
plt.show()
```
